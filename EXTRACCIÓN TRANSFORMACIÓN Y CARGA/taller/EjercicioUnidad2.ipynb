Formatos y colores para visualización de datos

```python
#Formato para la columna data science, Mes y machine learning
format_dict = {'data_science':'${:.2f}', 'Mes':'{:%d-%m-%Y}', 'machine_learning':'{:.2%}'}
#Poner formato de fecha a la columna Mes
df['Mes'] = pd.to_datetime(df['Mes'])
#Aplicar el estilo en la visualización
df.head().style.format(format_dict)
```

Resaltar máximos y mínimos del dataset con colores

```python
format_dict = {'Mes':'{:%d-%m-%Y}'} 
#Determina los valores máximos y mínimos con colores
df.head().style.format(format_dict).highlight_max(color='blue').highlight_min(color='pink')
```

Se va a utilizar una escala de colores creciente para mostrar los datos

```python
df.head(150).style.format(format_dict).background_gradient(subset=['data science', 'deep learning','categorical'], cmap='BuGn')
```

Visualizar los datos por medio de código de barras dentro de los datos

```python
df.head(150).style.format(format_dict).bar(color='green', subset=['data science','machine learning' ,'deep learning'])
```

```python
#En una presentación es posible combinar ambos filtros de datos para la visualización
df.head(10).style.format(format_dict).background_gradient(subset=['data science', 'machine learning'], cmap='BuGn').highlight_min(color='brown').highlight_max(color='pink')
```

Utilizando Pandas Profiling para reporte de datos

```python
import numpy as np
np.random.seed(0)
df2 = pd.DataFrame(np.random.randn(10,4), columns=['A','B','C','D'])
df2.style
```

Aplicación del comando applymap donde cambia el color de los valores negativos dentro de una tabla

```python
def style_negative(v, props=''):
    return props if v < 0 else None
s2 = df2.style.applymap(style_negative, props='color:blue;')\
              .applymap(lambda v: 'opacity: 20%;' if (v < 0.3) and (v > -0.3) else None)
s2
```

Función de determine los valores máximos dentro de la tabla

```python
def highlight_min(s, props=''):
    return np.where(s == np.nanmax(s.values), props, '')
s2.apply(highlight_min, props='color:white;background-color:red', axis=0)
```

```python
# Se puede usar esta función den diferentes columnas
s2.apply(highlight_max, props='color:white;background-color:red;', axis=1)\
  .apply(highlight_max, props='color:white;background-color:cyan', axis=None)
```

Diagrama de barras con diferentes condiciones por cada columna

```python
np.random.seed(0)
df2 = pd.DataFrame(np.random.randn(10,4), columns=['A','B','C','D'])
df2.style

df2.style.format('{:.3f}', na_rep="")\
         .bar(align=0, vmin=-2.5, vmax=2.5, cmap="bwr", height=50,
              width=60, props="width: 120px; border-right: 1px solid black;")\
         .text_gradient(cmap="bwr", vmin=-2.5, vmax=2.5)
```
